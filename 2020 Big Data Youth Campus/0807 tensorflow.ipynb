```python
from tensorflow.keras.wrappers.scikit_learn import KerasClassifier
```

```python
def build_fn():
    model = tf.keras.models.Sequential([
        tf.keras.layers.Dense(16, input_shape=(4,), activation = 'relu'),
        tf.keras.layers.Dense(16, activation = 'relu'),
        tf.keras.layers.Dense(3, activation = 'softmax')
    ])
    model.compile(loss = 'categorical_crossentropy', optimizer = 'adam', metrics = ['accuracy'])
    return model
```

```python
kc = KerasClassifier(build_fn, epochs = 10) # need build function
```

```python
kc.fit(iris.iloc[:,:-3], iris.iloc[:,-3:]) # epochs = 10를 여기다 안 넣고 위에 넣는다
```

```python
def build_fn():
    model = tf.keras.models.Sequential([
        tf.keras.layers.Dense(16, input_shape=(4,), activation = 'relu'),
        tf.keras.layers.Dense(16, activation = 'relu'),
        tf.keras.layers.Dense(10, activation = 'softmax')
    ])
    model.compile(loss = 'sparse_categorical_crossentropy', optimizer = 'adam', metrics = ['accuracy'])
    return model
```

```python
from tensorflow.keras.callbacks import Callback
```

```python
# overriding
Callback.on_epoch_begin # shift + tab
```

```python
class MyCallback(Callback):
    def on_epoch_begin(self, epoch, logs = None): # 상속
        print(epoch, 'start!!!')
```

```python
(X_train, y_train), (X_test, y_test) = tf.keras.datasets.mnist.load_data()
```

```python
X_train = X_train.reshape(-1, 784).astype("float32") / 255.0
```

```python
model.fit(X_train, y_train, epochs=5, callbacks = [MyCallback()])
```
