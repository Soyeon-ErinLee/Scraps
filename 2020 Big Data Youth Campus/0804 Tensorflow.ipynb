# Keras 실전 예제
- https://www.tensorflow.org/tutorials/quickstart/beginner

```python
(X_train, y_train), (X_test, y_test) = mnist.load_data()
```

```python
# 전처리
X_train, X_test = X_train / 255.0, X_test / 255.0 
```

```python
model = tf.keras.models.Sequential # 단방향모델 -> multi input, multi output이 안됨. 상속해서 쓰는 정도로 쓰임.
```

```python
model1 = tf.keras.models.Model 
```

```python
issubclass(model, model1) # 상속해서 만듦
```

```python
# https://keras.io/api/layers/core_layers/dense/
model = tf.keras.models.Sequential([
    tf.keras.layers.Dense(64, input_shape = (28,28)),  # Dense - fully connected layer
    tf.keras.layers.Dense(64),
    tf.keras.layers.Dense(10)
])
```

```python
model.summary()
```

```python
# flatten 시켜줘야 함
model = tf.keras.models.Sequential([
    tf.keras.layers.Flatten(input_shape = (28,28)),  # Flatten해서 전처리해주기
    tf.keras.layers.Dense(64),
    tf.keras.layers.Dense(10, name = 'output')
], name = 'moon')
```

```python
model.summary()
```

```python
# http://playground.tensorflow.org/
```

```python
# python의 weight -> tensorflow의 kernel, tensorflow의 weight -> kernel + bias 
```

```python
model.layers[1].weights
```

```python
# keras에서 function loss 지정하기
```

```python
tf.keras.losses.  # tab 누르면 온갖 losses 나온다
```

```python
model.compile(loss = 'sparse_categorical_crossentropy', optimizer = 'adam', metrics = ['accuracy'])
```

```python
model.fit(X_train, y_train, epochs=4)
```

```python
model.evaluate(X_test, y_test)
```
