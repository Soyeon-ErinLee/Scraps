```python
u_data = pd.read_csv('./data/ml-100k/u.data', header=None, sep='\t',
                    names=['user id' , 'item id' , 'rating' , 'timestamp'])
```

```python
u_data
```

```python
recommendation = u_data.copy()
```

```python
recommendation.drop(columns='timestamp', inplace=True)
```

```python
recommendation.set_index(['user id', 'item id']).unstack().fillna(0) # SCI (S)스택은 (C)컬럼을 (I)인덱스로
```

```python
rec_basic = recommendation.pivot(index='user id', columns='item id', values='rating').fillna(0)
```

```python
rec_corr = rec_basic.T.corr()
```

```python
rec_corr
```

```python
rec_corr.loc[730].sort_values(ascending=False)[1:6]
```

```python
def close_persons(user_id, number=5):
    return rec_corr.loc[user_id].sort_values(ascending=False)[1:number+1].index
```
