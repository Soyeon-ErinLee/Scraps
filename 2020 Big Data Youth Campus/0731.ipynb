```python
import seaborn as sns
```

```python
iris = sns.load_dataset('iris')
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(iris.iloc[:,:-2],iris.species)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier()
```

```python
knn.fit(X_train, y_train)
```

```python
knn.score(X_train, y_train)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
lr.score(X_test, y_test)
```

```python
from sklearn.model_selection import learning_curve
```

```python
learning_curve(KNeighborsClassifier(), iris.iloc[:,:-1], iris.species, cv=10)
```

```python
train_size, train_score, test_score = learning_curve(KNeighborsClassifier(), iris.iloc[:,:-1], iris.species, cv=10)
```

```python
import sklearn_evaluation
```

```python
# under/overfitting 인 지 알아냄
sklearn_evaluation.plot.learning_curve(train_score, test_score, train_size) # score과 size 순서 바뀌여야함 
```
